```python
''' from preprocessing class we import label encoder to convert
string data or categorical data into numerical'''

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()#initializing LabelEncoder()

df['Sex']=le.fit_transform(df['Sex'])
df['BP']=le.fit_transform(df['BP'])
df['Cholesterol']=le.fit_transform(df['Cholesterol'])
df['Drug']=le.fit_transform(df['Drug'])
```

```python
df.head()
```

```python
#defining dependent and independent valiable
x=df.iloc[:,0:5] #independent variable
x
```

```python
y=df.iloc[:,5:].values#dependent variable
```

```python
y
```

```python
#using standard scaler to get all the numerical value into a particular range
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler() #initializing the StandardScaler()
```

```python
s_x=ss.fit_transform(x)
```

```python
s_x
```

```python
# spliting dataset into training and testing set for model building
from sklearn.model_selection import train_test_split
```

```python
'''declaring training and testing variable with test size
as 30% where as training size will be 70% also we set ,
random state to 42 for more entropy through out the data'''

xtrain,xtest,ytrain,ytest=train_test_split(s_x,y,test_size=0.3,random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
#using tenser flow to built ANN model

'''here import sequential and dense from models and layars(class)
to add layers to ANN model'''

from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
ann=Sequential()#initializing Sequential()
```

```python
ann.add(Dense(5,activation='relu',kernel_initializer='random_uniform')) #input layer
```

```python
# Three hidden layers

ann.add(Dense(10,activation='relu',kernel_initializer='random_uniform'))
ann.add(Dense(15,activation='relu',kernel_initializer='random_uniform'))
ann.add(Dense(20,activation='relu',kernel_initializer='random_uniform'))
```

```python
#output layers
ann.add(Dense(5,activation='softmax',kernel_initializer='random_uniform')) #\
```

```python
#copiling model with optimizer as adam and metrics set to accuracy
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#here we fit the  model onto training set
ann.fit(xtrain,ytrain,batch_size=30,epochs=400)
```

```python
#here we made prediction on xtest split of data set and predict y test
pred=ann.predict(xtest)
```
